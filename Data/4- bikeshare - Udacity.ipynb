```python
# This is the my project in udacity.
# first we will import the important libraries in this project.
import time
import pandas as pd
import numpy as np

# Get the data
CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }

# use the functions to control the output.

# first function is used to filter the user input to match with the data
def get_filters():
        
    print('Hello! Let\'s explore some US bikeshare data!')
    
    #Using while loop to select the right city ,month and day listed
    while True:
        city=input("select one of these cities (chicago,new york city,washington): ").lower()
        month=input("select one of these months (Jan,Feb,Mar,Apr,May,Jun, or All): ").title()
        day=input("select one of these days (Saturday,Sunday,Monday,Tuesday,Wednesday,Thursday,Friday or All): ").title()
        try:
            if city in ("chicago" ,"new york city" ,"washington") and month in ('Jun', 'Mar', 'May', 'Apr','Jan','Feb','All') and day in ('Wednesday', 'Saturday', 'Tuesday', 'Sunday','Monday','Thursday',"Friday","All"):
                break
            else:
                print("Wrong input")
        except ValueError:
             print("Your input is wrong")

    print('-'*40)
    return city, month, day

# second we will load the data and to be matched with the user input.

def load_data(city, month,day):
    # Load the main data
    df1=pd.read_csv(CITY_DATA[city])

    # Use if to filter the output of the month to be matched with the input of the user.
    if month != "All":
        df_m=df1[pd.to_datetime(df1["Start Time"]).dt.strftime("%b")==month]
    else:
        df_m = df1
    
    # Use if to filter the output of the day to be matched with the input of the user.
    if day != "All":
        df=df_m[pd.to_datetime(df_m["Start Time"]).dt.day_name()==day]
    else:
        df=df_m

    return df

def time_stats(df):
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # display the most common month
    print("the most common month is: ")
    print(pd.to_datetime(df["Start Time"]).dt.strftime("%b").mode())

    # display the most common day of week
    print('the most common day of week is: ')
    print(pd.to_datetime(df["Start Time"]).dt.day_name().mode())

    # display the most common start hour
    print('the most common start hour is: ')
    print(pd.to_datetime(df["Start Time"]).dt.strftime("%H").mode())

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)



def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # display most commonly used start station
    print('the most commonly used start station is: ')          
    print(df["Start Station"].mode())

    # display most commonly used end station
    print('the most commonly used end station is: ')
    print(df["End Station"].mode())

    # display most frequent combination of start station and end station trip
    group_field=df.groupby('Start Station')['End Station'].value_counts().sort_values(ascending=False)
    print('the most frequent combination of start station and end station trip are: ')
    print(pd.DataFrame(group_field.head(1)))

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # display total travel time
    print('the total travel time is: ')
    print(sum(df["Trip Duration"]))

    # display mean travel time
    print('the mean of travel time is: ')
    print(df["Trip Duration"].mean())

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def user_stats(df,city):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()

    # Display counts of user types
    print('the counts of user types is: ')
    print(df["User Type"].count())

    # Display counts of gender
    
    if city == "washington" :
        print("There is no gender in this country")
    else:
        print('the counts of gender is: ')
        print(df["Gender"].count())
        
    # Display earliest, most recent, and most common year of birth
    if city == "washington" :
        print("There is no Birth Year in this country")
        
    else:
        print('the most common year of birth is: \n',df["Birth Year"].min(),"\n",'The earliest year is: \n',"\n",df["Birth Year"].max(),'the older year is: \n',df["Birth Year"].mode())

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)



def main():
    while True:
        
        city, month, day = get_filters()
        df = load_data(city, month, day)
        
        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df,city)
        
        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
    main()
```
