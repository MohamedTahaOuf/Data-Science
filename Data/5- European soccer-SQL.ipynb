### Research Question 1 : Which players have the most penalties?

```python
pd.read_sql('''select pa.player_api_id , p.player_name ,sum(pa.penalties)
                from Player_Attributes as pa
                inner join Player as p
                on pa.player_api_id = p.player_api_id 
                group by pa.player_api_id
                order by sum(pa.penalties) desc;''',con)[0:10]
```

<a id='Q2'></a>

### Research Question  2- The team who has the higher goals.

```python
team_score=pd.read_sql('''select match.season , match.home_team_api_id , team.team_long_name  ,
                sum(match.home_team_goal)+sum(match.away_team_goal) as Sum_of_goals
                from match
                inner join Team
                on match.home_team_api_id = team.team_api_id
                group by match.home_team_api_id
                order by Sum_of_goals desc
                ;''',con)
team_score
```

### Research Question  3- The average of speed,passing and crossing team

```python
x=pd.read_sql('''with uni as ( 
select Team.team_long_name, Team_Attributes.team_api_id ,avg(Team_Attributes.buildUpPlaySpeed) , avg(Team_Attributes.buildUpPlayPassing),avg(Team_Attributes.chanceCreationPassing),avg(Team_Attributes.chanceCreationCrossing) ,count(Team_Attributes.team_api_id) as Count
from Team_Attributes
inner join Team
on Team_Attributes.team_api_id = Team.team_api_id
group by Team_Attributes.team_api_id
order by avg(Team_Attributes.buildUpPlaySpeed) desc)
select * from uni
where Count >4
''',con)
x
```

### Research Question  4- The Premier league Ranking 2008/2009


```python
# Here we gonna use SQL to figure out the ranking of the Premier league in 2008/2009
df=pd.read_sql('''SELECT Team.team_long_name As home_team_long_name,
match.home_team_api_id,
match.home_team_goal,
(CASE
WHEN home_team_goal = away_team_goal THEN 1
WHEN home_team_goal > away_team_goal THEN 3
WHEN home_team_goal < away_team_goal THEN 0
END)Home_Points,

OpponentTeam.team_long_name AS away_team_long_name,
match.away_team_api_id,
match.away_team_goal,
(CASE
WHEN home_team_goal = away_team_goal THEN 1
WHEN home_team_goal < away_team_goal THEN 3
WHEN home_team_goal > away_team_goal THEN 0
END)Away_Points

FROM match 
JOIN Country ON match.country_id = Country.id
JOIN Team ON match.home_team_api_id = Team.team_api_id 
JOIN Team AS OpponentTeam ON match.away_team_api_id = OpponentTeam.team_api_id
WHERE season='2008/2009' AND Country.name='England'
''',con)
df
```

```python
#after that we will use concatenate in pandas to combain the columns and then using groupby to find the highest point and goals
concat=pd.DataFrame(np.concatenate((df[["home_team_long_name","Home_Points","home_team_goal"]],
                                    df[["away_team_long_name","Away_Points","away_team_goal"]]), axis=0))
concat.columns =["Team_Name","Points","Goals"]
concat
pd.DataFrame(concat.groupby("Team_Name")["Goals","Points"].sum()).sort_values(by='Points', ascending=False)
```
